##  Download and Prep the Data: 1 Mark

<h4>Import the libraries needed</h4>

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

<h4> Load data and store in dataframe </h4>

```python
url = "https://drive.google.com/uc?export=download&id=1ZQCx34-PvKLuQFWlKyeamM8Un2liJTQ4"
car_dataframe = pd.read_csv(url)
car_dataframe
```

<h4>Find out  what type of variable you are dealing with. This will help you find the right visualization method for that variable.</p>


```python
# Separate numerical and categorical variables
numerical_vars = car_dataframe.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_vars = car_dataframe.select_dtypes(include=['object']).columns.tolist()

# Print variable types
print("Numerical Variables:", numerical_vars)
print("Categorical Variables:", categorical_vars)
```

<h4> Find the correlation between the following columns: bore, stroke, compression-ratio, and horsepower.<br>
Write down your observations in a markdown cell below. <h4>

```python
# Correlation analysis
correlation_cols = ["bore", "stroke", "compression-ratio", "horsepower"]
correlation_matrix = car_dataframe[correlation_cols].corr()

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix: Bore, Stroke, Compression Ratio, and Horsepower")
plt.show()
```

**Observations**
- **Bore and Horsepower** are in **strong positive correlation (0.57)**, i.e., increasing bore size increases horsepower.
- **Stroke and Compression Ratio** demonstrate a **weak positive correlation (0.19)**.
- **Compression Ratio and Horsepower** exhibit a **negative correlation (-0.21)**, indicating that higher compression ratios could result in lower horsepower.
- **Bore and Stroke** correlate nearly zero (-0.06), meaning these factors are quite independent.

```python
# Visualizing correlations between price and key variables
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Scatter plot for Engine Size vs Price
sns.scatterplot(x=car_dataframe['engine-size'], y=car_dataframe['price'], ax=axes[0], color='red')
axes[0].set_title("Engine Size vs Price")
axes[0].set_xlabel("Engine Size")
axes[0].set_ylabel("Price")

# Scatter plot for Highway MPG vs Price
sns.scatterplot(x=car_dataframe['highway-mpg'], y=car_dataframe['price'], ax=axes[1], color='blue')
axes[1].set_title("Highway MPG vs Price")
axes[1].set_xlabel("Highway MPG")
axes[1].set_ylabel("Price")

# Scatter plot for Peak RPM vs Price
sns.scatterplot(x=car_dataframe['peak-rpm'], y=car_dataframe['price'], ax=axes[2], color='green')
axes[2].set_title("Peak RPM vs Price")
axes[2].set_xlabel("Peak RPM")
axes[2].set_ylabel("Price")

plt.tight_layout()
plt.show()
```

### Observations:
1. **Engine Size vs Price**: A **strong positive correlation**. Price goes up significantly as engine size goes up. **Engine size is the best predictor variable** for price out of the three.
2. **Highway MPG vs Price**: A **strong negative correlation**. Higher fuel efficiency (MPG) vehicles are cheaper.
3. **Peak RPM vs Price**: Exhibits **little to no correlation**, i.e., peak RPM has little to no effect on car prices.

### Pre-attentive Attributes Employed:
- **Color**: Various colors employed in scatter plots to differentiate relationships.
- **Position**: Points are positioned in such a manner that trends are clearly exhibited.

### Gestalt Principles Employed:
- **Proximity**: Points which are near each other signify stronger relationships.
- **Similarity**: Various colors assist in differentiating various variables in the scatter plots.

```python

# Creating Pivot Table for Heatmap
df_pivot = car_dataframe.pivot_table(values='price', index='drive-wheels', columns='body-style', aggfunc='mean')

# Plot Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df_pivot, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Heatmap of Price by Drive-Wheels and Body-Style")
plt.xlabel("Body Style")
plt.ylabel("Drive Wheels")
plt.show()
```

### Observations (Heatmap):

 <h4>How does engine size and curb weight together influence the price of a car?<br><br>

```python
# Scatter plot for Engine Size and Curb Weight vs Price
plt.figure(figsize=(10, 6))
scatter = plt.scatter(car_dataframe['engine-size'], car_dataframe['curb-weight'], c=car_dataframe['price'], cmap='coolwarm', s=car_dataframe['price']/100, alpha=0.75)
plt.title("Engine Size and Curb Weight vs Price")
plt.xlabel("Engine Size")
plt.ylabel("Curb Weight")
plt.colorbar(scatter, label='Price')
plt.show()
```

### Observations:
1. **Effect of Engine Size and Curb Weight on Price**: Large engine sizes and higher curb weight generally correspond with higher prices.
2. **Rationale Behind the Choice of Chart**: The use of color and size changes in a scatterplot clearly represents the combined impact of two continuous variables on price.
3. **Pre-attentive Features Employed:**
- **Color**: Represents changes in price and allows high and low prices to be easily distinguishable.
- **Size**: The larger points represent higher prices, so the pattern is self-explanatory.
4. **Gestalt Principles Applied:**
- **Proximity**: Prices within similar ranges tend to group together naturally.
- **Continuity**: The trend from engine size, curb weight, to price is seen through continuous transitions in color and size.
